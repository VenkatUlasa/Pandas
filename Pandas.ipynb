##Pandas :

We use Pandas in data science and data analysis because it provides powerful tools for working with structured data. It simplifies data manipulation, analysis, and cleaning tasks, making it easier to handle data from various sources (like Excel, CSV files, SQL databases, etc.). Here are the key reasons why Pandas is so widely used:



*   Series:
 A one-dimensional array-like structure, similar to a list, but with labeled indices.
*   DataFrame:
A two-dimensional, tabular data structure with labeled axes (rows and columns), similar to a table in a relational database or a spreadsheet. Itâ€™s the core data structure in Pandas for working with datasets.

```python
# Reading CSV file using URL : we can use file path if the file is downloaded ;
# Load Titanic dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df
```

```python
# head() => returns specified no of rows from the top of the data set;

df.head() # by default it returns 5 rows and we can specify no of rows we want to print ;
```

```python
# tail() => returns specified no of rows from the bottom of the data set;

df.tail(10) # by default it returns 5 rows and we can specify no of rows we want to print ;
```

```python
# sample() => this is same like head() but it returns the rows randomly from given dataset;
df.sample(10) # By default it returns 1 row but we can specify ;
```

```python
df.shape
```

```python
# info() => it returns complete information about the given dataset;
df.info()
```

```python
df.isnull().sum()
```

```python
# removing some unwanted columns from dataset;
df = df.drop(['PassengerId','Name','Ticket'],axis = 1)  # it removes the column is existed otherwise it returns an error.
```

```python
df.head()
```

```python
# unique()
df["Sex"].unique()
```

```python
# value_counts() => it returns the total count of each unique value from specified column.

df['Sex'].value_counts()
```

```python
# filling null values with a value.
# fillna()
df["Age"] = df["Age"].fillna(10)
```

```python
df["Age"].isnull().sum()  # returns 0 because all null values filled with 10;
```

```python
df.head()
```

```python
df.shape
```

```python
# removing rows having null values;
# dropna()
df = df.dropna()
```

```python
df.shape # we removed all the rows having null values;
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# all the columns with only numerical values;

num_cols = df.select_dtypes(exclude='object')  # exclude='object' means i don't need columns with object(text) data type;
```

```python
num_cols # it returns only columns with numerical values;
```

```python
# now columns with text values;

txt_cols = df.select_dtypes(include = "object")
```

```python
txt_cols
```

```python
# we need to add two data sets side by side ;
# concat() it is the function used to add multiple data sets with having equal number of rows side by side;
n_ds = pd.concat([num_cols,txt_cols],axis = 1)
n_ds
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

```python
# map()

df["Sex"] = df["Sex"].map({'male':1,'female':2}).astype(int)
df.head()
```

```python
df = df.drop(["PassengerId","Name","Ticket"],axis = 1)
```

```python
df.head()
```

```python
# We can find the mean() , median() and mode() for the columns having only numerical values.
# in this i am taking the column "Fare"

print(df["Fare"].mean()) # prints mean value.
print(df["Fare"].median())  # prints median value.
print(df["Fare"].mode()) # prints mode value.
```
